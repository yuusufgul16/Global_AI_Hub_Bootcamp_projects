Which year has the greatest total runtime?


---

# **START**

```python
year_max_runtime = data
year_max_runtime["Premiere"] = pd.to_datetime(year_max_runtime.Premiere)
year_max_runtime["Year"] = year_max_runtime.Premiere.dt.year
year_max_runtime = year_max_runtime.groupby("Year")["Runtime"].sum()
year_max_runtime = year_max_runtime.reset_index()
year_max_runtime.columns = ["Year","Runtime"]
plt.bar(year_max_runtime.Year,year_max_runtime.Runtime)
plt.show()
```
